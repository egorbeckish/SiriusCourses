```python
# n школьников делят между собой k яблок поровну, неделящийся остаток остаётся в корзинке.
# Сколько яблок достанется каждому школьнику? Сколько яблок останется в корзинке?

# Программа получает на вход в первой строке натуральное число n,
# а во второй — целое неотрицательное число k и должна вывести два целых числа:
# количество яблок у каждого школьника и количество яблок, оставшихся в корзинке.

n = int(input())
k = int(input())

print(k // n, k % n)
```

```python
# Феофилакт хочет купить ракетки и шарики для игры в настольный теннис.
# Один комплект ракеток стоит A рублей, один шарик стоит B рублей.
# У Феофилакта есть C рублей, C ≥ A, он покупает один комплект ракеток и шарики на оставшиеся деньги.
# Сколько шариков купит Феофилакт?

# Программа получает на вход три целых числа: A, B, C. Программа должна вывести ответ на задачу.

A = int(input())
B = int(input())
C = int(input())

print((C - A) // B)
```

```python
# Дано натуральное число. Найдите число десятков в его десятичной записи (то есть предпоследнюю цифру его десятичной записи).
# Если заданное число является однозначным, то необходимо вывести 0.

n = int(input())
print((n % 100) // 10)
```

```python
# Дано четырёхзначное число. Найдите сумму его цифр.

n = int(input())

s = 0
while n != 0:
    s += n % 10
    n //= 10

print(s)
```

```python
# С начала некоторых суток прошло n минут.
# Определите, сколько часов и минут будут показывать электронные часы в этот момент.

# На вход программе подаётся целое неотрицательное число n.

# Программа должна вывести два числа: количество часов (от 0 до 23) и количество минут (от 0 до 59).

n = int(input())

print(n // 60, n % 60)
```

```python
# Из проволоки толщиной d миллиметров сделали кольца.
# Внутренний радиус каждого кольца составляет R миллиметров.
# Всего сделали n колец, и их соединили в цепь. Определите длину получившейся цепи L.
# На рисунке изображён пример для n = 3.

# Входные данные

# Программа получает на вход три числа, записанных в отдельных строках.
# В первой строке задана толщина проволоки d (в миллиметрах).
# Во второй строке задан внутренний радиус кольца R (в миллиметрах).
# В третьей строке задано число звеньев n.

# Выходные данные

# Программа должна вывести одно целое число L — суммарную длину получившейся цепи.

d = int(input())
R = int(input())
n = int(input())

L = 2 * (n * R + d)
print(L)
```

```python
# Пирожок в столовой стоит a рублей и b копеек. Определите, сколько рублей и копеек нужно заплатить за n пирожков.

# Программа получает на вход три числа a, b и n и должна вывести два числа: стоимость покупки в рублях и копейках.

a = int(input())
b = int(input())
n = int(input())

print(a * n + (b * n) // 100, (b * n) % 100)
```

```python
# n школьников делят k яблок «почти поровну», то есть так, чтобы количество яблок,
# доставшихся любым двум школьникам, отличалось бы не более, чем на 1.

# Программа получает на вход натуральное число n и целое неотрицательное число k,
# каждое в отдельной строке, и должна вывести количество школьников,
# которым достанется яблок меньше, чем некоторым из их товарищей.


n = int(input())
k = int(input())

print((n - k % n) % n)
```

```python
# За день машина проезжает n километров. Сколько дней нужно, чтобы проехать маршрут длиной m километров?

# Программа получает на вход в первой строке натуральное число n и во второй строке — целое неотрицательное число m.

n = int(input())
m = int(input())

print((m + n - 1) // n)
```

```python
# В некоторой школе решили набрать три новых математических класса и оборудовать кабинеты для них новыми партами.
# За каждой партой могут сидеть двое учащихся. Известно количество учащихся в каждом из трёх классов.
# Определите наименьшее число парт, которое нужно приобрести для них.
# Обратите внимание, что школьники из разных классов не могут сидеть за одной партой.

# Программа получает на вход три неотрицательных целых числа:
# количество учащихся в каждом из трёх классов, каждое в отдельной строке.
# Программа должна вывести ответ на задачу.

a = int(input())
b = int(input())
c = int(input())

s = a // 2 + a % 2 + b // 2 + b % 2 + c // 2 + c % 2
print(s)
```

```python
# На каждой странице книги напечатано ровно k строк:
# на первой странице находятся строки с 1 по k, на второй — с k + 1 по 2k и т. д.
# Определите, на какой странице находится строка номер n и какой по счёту будет эта строка на странице.

# Даны натуральные числа k и n, каждое в отдельной строке.
# Программа должна считать их и вывести два числа: номер страницы и номер строки на странице.

k = int(input())
n = int(input())

print((k + n - 1) // k, (n - 1) % k + 1)
```

```python
# Шахматная доска состоит из n × m клеток, покрашенных в чёрный и белый цвет в шахматном порядке.
# При этом клетка в левом нижнем углу доски покрашена в чёрный цвет. Определите, сколько всего на доске чёрных клеток.

# Программа получает на вход натуральные числа n и m. Программа должна вывести ответ на задачу.

n = int(input())
m = int(input())

print((n * m) // 2 + (n * m) % 2)
```

```python
# Дано целое число n. Выведите следующее за ним чётное число.

n = int(input())

a = n // 2
b = n % 2
c = n // a

print(n + (c - b))
```

```python
# Дано целое неотрицательное число, меньшее 10000. Если число имеет меньше 4 знаков в десятичной записи,
# то нужно считать, что его десятичная запись дополняется слева незначащими нулями
# (например, число 120 дополняется до 0120). Определите, является ли его десятичная запись симметричной.
# Если число симметричное, то выведите 1, иначе выведите любое другое целое число.

n = int(input())
a = n // 1000
b = (n // 100) % 10
c = (n // 10) % 10
d = n % 10
print((a - d) ** 2 + (b - c) ** 2 + 1)
```

```python
# Напишите программу, которая считывает два натуральных числа a и b и выводит наибольшее значение из них.

# При решении задачи можно пользоваться только целочисленными арифметическими операциями +, −, *, //, %, =.
# Нельзя пользоваться такими конструкциями, как ветвления, циклы, функции.

a = int(input())
b = int(input())

print(int((a + b + ((a - b) ** 2) ** .5) // 2))
```
